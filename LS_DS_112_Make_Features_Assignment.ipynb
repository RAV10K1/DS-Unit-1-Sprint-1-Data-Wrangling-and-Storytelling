```python
# Loading dataset into dataframe and checking head
df = pd.read_csv('LoanStats_2018Q4.csv', header=1, skipfooter=2, engine='python')
df.head()
```

```python
# Checking shape of dataframe
df.shape
```

```python
# Checking for null values
pd.set_option('display.max_rows', 200)
df.isnull().sum().sort_values(ascending=True)
```

```python
# Dropping NaN columns
df = df.drop(['url', 'member_id', 'desc', 'id'], axis=1)
```

```python
df.shape
```

```python
# Checking interest rate column
df['int_rate'].sample(10)
```

```python
# Creating function to convert object to float in interest rate column
def cleaner(cell_contents):
  return float(cell_contents.strip().strip('%'))
```

```python
# Applying function to entire interest rate column
new_list = []
for cell in df['int_rate']:
  new_list.append(cleaner(cell))
```

```python
# Overwriting interest rate column in dataframe
df['int_rate'] = pd.Series(new_list)
df.head()
```

```python
# Checking data type of overwritten column
df['int_rate'].dtypes
```

```python
# Checking for NaN values in emp_title
df['emp_title'].isnull().value_counts()
```

```python
# Checking top 20 employment titles
df['emp_title'].value_counts(dropna=False)[:20]
```

```python
# Checking unique employment titles
len(df['emp_title'].unique())
```

```python
# Creating list of corrected job titles
job_titles = ['owner', 'Supervisor', ' Project Manager', np.nan]
```

```python
# Defining function to clean the job titles
def clean_title(job_title):
  if isinstance(job_title, str):
    return job_title.title().strip()
  else:
    return "Unknown"
```

```python
# Creating a for loop for the function
empty_list = []
for title in job_titles:
  empty_list.append(clean_title(title))
```

```python
# Using list comprehension
[clean_title(title) for title in job_titles]
```

```python
# Overwriting job titles into relevant column of the data frame
df['emp_title'] = df['emp_title'].apply(clean_title)
df.head()
```

```python
# Checking for NaN values
df['emp_title'].isnull().sum()
```

```python
# Checking for Date Columns
my_cols = [col for col in df.columns if col.endswith('_d')]
my_cols
```

```python
# Checking formats of the dates in issue_d
df['issue_d'][:10]
```

```python
# Inferring date-time for column
df['issue_d'] = pd.to_datetime(df['issue_d'],infer_datetime_format=True)
df[my_cols].head()
```

```python
# Using .dt accessor
df_cols = df[my_cols].copy()
```

```python
# Creating new column for year
df_cols['year'] = df_cols['issue_d'].dt.year
df_cols.head()
```

```python
# Creating new column for month
df_cols['month'] = df_cols['issue_d'].dt.month
df_cols.head()
```

```python
# Inferring date-time for last_credit_pull_date
df_cols['last_credit_pull_d'] = pd.to_datetime(df_cols['last_credit_pull_d'], infer_datetime_format=True)
df_cols.head()
```

```python
# Creating new column for difference in days
df_cols['day_diff'] = (df_cols['last_credit_pull_d'] - df_cols['issue_d']).dt.days
df_cols.head()
```

```python
df_cols['day_diff'].sort_values(ascending=False)[:20]
```

```python
# Function for removing months in term column
def month_clean(cell_contents):
  return float(cell_contents.strip().strip('months'))
```

```python
# Applying function to term column
df['term'] = df['term'].apply(month_clean)
```

```python
# Cleaning revol_util column by removing %
df['revol_util'] = df['revol_util'].str.rstrip('%')
```

```python
# Converting colum to float
df['revol_util'] = df['revol_util'].astype(float)
```

```python
# Modifying emp_title column to depict top 20 titles

df['emp_title'].value_counts().sort_values(ascending=False)[:21]
```

```python
# Cleaning emp_title
df['emp_title'] = df['emp_title'].replace(to_replace='Rn', value='Registered Nurse')
```

```python
df['emp_title'].value_counts().sort_values(ascending=False)[:21]
```

```python
# Defining function to categorize emp_titles not in top 20 as 'Other' 

top_titles = ['Teacher','Unknown','Manager','Registered Nurse','Driver','Supervisor', 'Truck Driver', 
               'Office Manager','Sales', 'Project Manager', 'Owner', 'Director', 'Operations Manager', 
               'Sales Manager', 'Police Officer', 'Nurse', 'Technician', 'Engineer', 'Administrative Assistant', 'Store Manager']

def emp_title_clean(cell_contents):
  if isinstance(top_titles, str):
    return cell_contents
  else:
    return 'Other'
```

```python
# Applying function to relevant column
df['emp_title'] = df['emp_title'].apply(emp_title_clean)
```
